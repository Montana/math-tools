# Defining terms

*Polynomial* One of those equations with coefficients, variables, and non negative exponentiation combined with addition/multiplication/etc.

*Quadratic* functions are polynomials of degree 2.

*Parabola* - graph of a quadratic.  They have those open face shapes and point up or down. 

One definition of a parabola is that it's the set of points which are equidistant from a Focus (a point) and a Directrix (Line).

Forms of equations for a parabola with examples:

- *Standard Form* - $ 3x^2 + 12x - 15 $
- *Factored Form* - $ 3(x-1)(x+5) $
- *Vertex Form* - $ 3(x+2)^2 - 27 $

Abstraction of vertex form we learned in Geometry lesson:
$$ y = a(x-h)^2 + k $$
$$ x = a(y-k)^2 + h $$

Where $ (h, k) $ is the *Vertex* of the parabola. That Vertex is either the top or the bottom.

Aha! Just noticed the Vertex and the Root (derived from factored form) could be same, if there's no constant term.

*Monic* quadratics are expressions where the leading coefficient ($a$) is equal to $1$.

# Completing the Square

Completing the square works on any quadratic, and is actually the proof basis of the Quadratic Formula.

Make sure to divide out coefficient $a$ before completing the square.  Then just add a constant factor thats the square of half coefficient $b$ to do your squaring proper.

Completing the square works cus it lets you take a square root of both sides to eliminate the 2nd degree factor.

Make sure that when you complete the square and do the final square root step, that you remember this produces a $ \pm $ value.

# Quadratic Formula

$ x=\frac{-b\pm\sqrt{b^2-4ac}}{2a} $

The discriminant is the part of the quadratic formula underneath the square root symbol: $ b^2-4ac $. The discriminant tells us whether there are two solutions, one solution, or no solutions.

* Positive Discriminant - $ b^2 - 4ac > 0 $ - 2 real roots
* Zero Discriminant - $ b^2 - 4ac = 0 $ - 1 repeated real 0 root.
* Negative Discriminant - $ b^2 - 4ac < 0 $ - 2 imaginary roots.

## Proof of Quadratic Formula using Completing the Square

$$
\begin{align*}
ax^2 + bx + c = 0 \\
x^2 + \frac{bx}{a} + \frac{c}{a} = 0 \\
x^2 + \frac{bx}{a} + \frac{c}{a} + ((\frac{b}{2a})^2 - \frac{c}{a}) = ((\frac{b}{2a})^2 - \frac{c}{a}) \\
x^2 + \frac{bx}{a} + (\frac{b}{2a})^2  = (\frac{b}{2a})^2 - \frac{c}{a} \\
(x + \frac{b}{2a})^2 = (\frac{b}{2a})^2 - \frac{c}{a} \quad\text{Complete the Square} \\
(x + \frac{b}{2a})^2 = \frac{b^2}{4a^2} - \frac{c}{a} \\
(x + \frac{b}{2a})^2 = \frac{b^2 - 4ac}{4a^2} \\
x + \frac{b}{2a} = \pm\sqrt{\frac{b^2 - 4ac}{4a^2}} \\
x + \frac{b}{2a} = \frac{\pm\sqrt{b^2 - 4ac}}{2a} \\
x = \frac{-b\pm\sqrt{b^2 - 4ac - b}}{2a}
\end{align*}
$$
